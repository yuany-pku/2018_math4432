```python
# 4 Embarked
for dataset in fulldata:
    dataset['Embarked'] = dataset['Embarked'].fillna('S')

# The same for 5 Fare
for dataset in fulldata:
    dataset['Fare'] = dataset['Fare'].fillna(train['Fare'].median())
train['CategoricalFare'] = pd.qcut(train['Fare'], 4)
#6 Age

```

```python
# 7 Name
def get_title(name):
	title_search = re.search(' ([A-Za-z]+)\.', name)
	# If the title exists, extract and return it.
	if title_search:
		return title_search.group(1)
	return ""

for line in fulldata:
    line['Title'] = line['Name'].apply(get_title)

for dataset in fulldata:
    dataset['Title'] = dataset['Title'].replace(['Lady', 'Countess','Capt', 'Col',\
 	'Don', 'Dr', 'Major', 'Rev', 'Sir', 'Jonkheer', 'Dona'], 'Rare')

    dataset['Title'] = dataset['Title'].replace('Mlle', 'Miss')
    dataset['Title'] = dataset['Title'].replace('Ms', 'Miss')
    dataset['Title'] = dataset['Title'].replace('Mme', 'Mrs')

print(pd.crosstab(traindf['Title'], traindf['Sex']))
print (traindf[['Title', 'Survived']].groupby(['Title'], as_index=False).mean())
```

```python
# Drop List
drop_col = ['PassengerId','Name']
test_id = testdf.PassengerId

train = traindf.drop(drop_col,axis = 1)
test = testdf.drop(drop_col,axis = 1)
train.head()
test.head()
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
train.head()
lbl_array = []
one_array = []
catarray = ['Sex', 'Ticket', 'Cabin', 'Embarked', 'Title']
for col in catarray:
    lbl = LabelEncoder()
    lbl.fit(list(train[col].values)+(list(test[col].values))  )
    train.loc[:,col] = lbl.transform(list(train[col].values))
    test.loc[:,col] = lbl.transform(list(test[col].values))
    lbl_array.append(lbl)
#     one = OneHotEncoder()
#     one.fit(list(train[col].values))
#     train.loc[:,col] = one.transform(list(train[col].values))
#     one_array.append(one)
train.head()
test.head()
```

```python
# Using XGB to train and test
# encode_label = ['Pclass','Sex','SibSp']
# lbl = preprocessing.LabelEncoder()
# lbl.fit(list(df))
import xgboost as xgb
params = {
        'objective': 'binary:logistic',
        'eta': 0.1,
        'max_depth': 9,
        'seed': 10,
        'missing': -999,
        'silent' : 0
}
xgbtrain = xgb.DMatrix(train.drop(['Survived'],axis=1),train.Survived)
xgbtest = xgb.DMatrix(test)
watchlist = [(xgbtrain,'train'),(xgbtest,'test')]
num_rounds = 5000
model = xgb.train(params, xgbtrain, num_rounds)
```

```python
train_pre = model.predict(xgb.DMatrix(train.drop(['Survived'],axis=1)))
train_result = pd.DataFrame([train.Survived,train_pre]).T
train_result.columns = ['s','p']
train_result.head()
for i in range(1,10,1):
    (sum(train_result.p>(0.1*i)))/891
```

```python
test['Survived'] = model.predict(xgbtest)
test.head()
result = pd.DataFrame()
result['PassengerId'] = test_id
result['Survived'] = test.Survived>=0.5
# result.info()
result.loc[:,'Survived'] = result.Survived.apply(lambda x: int(x))
result.head()
result.to_csv('result.csv',index=False)
```
